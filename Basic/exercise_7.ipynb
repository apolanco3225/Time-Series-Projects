# White Noise

```python
import random
import pandas as pd
from pandas.plotting import autocorrelation_plot
import seaborn as sns
sns.set_style("darkgrid")
```

```python
random.seed(1)
series = [random.gauss(0.0, 1.0) for i in range(1000)]
```

```python
data = pd.Series(series)
```

```python
data.describe()
```

```python
data.plot()
```

```python
data.hist()
```

```python
autocorrelation_plot(data)
```
