# Power Transform

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import plotly.express as px
```

```python
df = pd.read_csv(
    "data/airline-passengers.csv", 
    index_col="Month", 
    parse_dates=["Month"]
)
df.head()
```

```python
df.info()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(df)
plt.subplot(212)
plt.hist(df);
```

```python
import numpy as np
```

# Squared Root Transform (Power Transform)

```python
squared_df = np.square(df)

plt.figure(1)
plt.subplot(211)
plt.plot(squared_df)
plt.subplot(212)
plt.hist(squared_df);
```

# Log Transform (Power Transform)

```python
log_df = np.log(df)

plt.figure(1)
plt.subplot(211)
plt.plot(log_df)
plt.subplot(212)
plt.hist(log_df);
```

# Box-Cox Transform (Configurable Data Transform)
- lambda = -1. is a reciprocal transform.
- lambda = -0.5 is a reciprocal square root transform.  
- lambda = 0.0 is a log transform.
- lambda = 0.5 is a square root transform.
- lambda = 1.0 is no transform.

```python
from scipy.stats import boxcox
```

```python
boxcox_df = boxcox(
    df.iloc[:, 0].values, 
    lmbda=0.0
)

plt.figure(1)
plt.subplot(211)
plt.plot(boxcox_df)
plt.subplot(212)
plt.hist(boxcox_df);
```

```python
boxcox_df, lambda_value = boxcox(
    df.iloc[:, 0].values, 
)
print(lambda_value)
plt.figure(1)
plt.subplot(211)
plt.plot(boxcox_df)
plt.subplot(212)
plt.hist(boxcox_df);
```
